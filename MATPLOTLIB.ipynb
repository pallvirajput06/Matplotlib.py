```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

MATPLOATLIB BASIC




```python
x = np.linspace(0,5,11)
x

y = x**2
y
```

```python
plt.plot(x,y)  #or plt.plot(y,x)
plt.xlabel('x axis')
plt.xlabel('y axis')
```

SUB PLOT IS MATPLOATLIB

```python
plt.subplot(2,2,1)#convert canva into 2 rows and 2 column, third ploat inside() stand for selecting subcolumn value
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x**2,x**2)
```

```python
plt.subplot(6,6,1)
```

OBJECT ORIENTED WAY

```python
fig = plt.figure()#for selecting canvas
fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure()
axis = fig.add_axes([0.1,0.1,0.6,0.8])
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.6,0.8])
axis1.plot(x,y)

axis2 = fig.add_axes([0.1,0.1,0.2,0.2])
axis.plot(x,y)
```

```python
#set fixed size and dpi
fig = plt.figure(figsize=(20,8) , dpi =100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
```

TYPES OF PLOT

```python
#1.scatter plot
plt.scatter(x,y)
```

```python
#2. Histogram plot
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
#3. boxplot
plt.boxplot(data)
```

Saving plot

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.6,0.8])
axis1.plot(x,y)

plt.savefig("Basicplot.png")
```
