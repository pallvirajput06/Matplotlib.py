```python
import matplotlib.pyplot as plt
```

Line Plot

```python
x = [1,2,3,4,5]
y = [10,20,25,40,50]
```

```python
plt.plot(x,y)
#foe labeling x ,y,title axis
#x,y must be of same length
plt.xlabel('This is x label')
plt.ylabel('This is y label')
plt.title('This is your line plot')

plt.show
```

Canvas

```python
#canvas length control
plt.figure(figsize=(10,10))
```

```python
plt.figure(figsize=(6,4))

x =[1,2,3,4,5]
y1 =[1,4,9,16,25]
y2 = [1,2,3,4,5]



#line 1 plot
plt.plot(x,y1)
plt.plot(x,y2)


#labels
plt.plot(x,y1,label = 'Line1')
plt.plot(x,y2,label = 'Line2')

plt.legend()#to show label we use legend

plt.grid()#for square canva

```
