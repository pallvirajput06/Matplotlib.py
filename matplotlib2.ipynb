```python
import matplotlib.pyplot as plt
```

SubPLOT

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,4)
plt.plot(x,y2)

plt.show()
```

```python
#controling axis of graph
#controlling graps canvas
#sample data
x = [1,2,3,4,5]
y = [2,4,6,8,10]
```

```python
fig = plt.figure(figsize=(6,4))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #[left,bottom,width,height]
axes1.plot(x,y)
plt.title('Mera phone')

axes2 = axes1 = fig.add_axes([0.2,0.5,0.3,0.3]) #[left,bottom,width,height]
axes1.plot(x,y)
plt.title('Remote')

plt.show()


```

```python
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,4,5,7]

plt.plot(x,y1, label = 'Nili line', color="black",lw=2.5,ls= '--')
plt.plot(x,y2, label = 'Orage line',marker='*')
plt.legend()
plt.show()
```
